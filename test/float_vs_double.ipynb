```python
import matplotlib.pyplot as plt
import math
```

```python
inpFloat, quadFloat = list(), list()
with open('iofiles/float_inphase_quadrature_output_gmsk_modulate.txt') as f: 
    lines = f.readlines()
    for i in range(0, len(lines), 3): 
        line1 = lines[i].strip()
        line2 = lines[i + 1].strip()
        inpFloat.append(float(line1))
        quadFloat.append(float(line2))
```

```python
inpDouble, quadDouble = list(), list()
with open('iofiles/double_inphase_quadrature_output_gmsk_modulate.txt') as F:
    lines = F.readlines()
    for i in range(0, len(lines), 3): 
        line1 = lines[i].strip()
        line2 = lines[i + 1].strip()
        inpDouble.append(float(line1))
        quadDouble.append(float(line2))
print(inpDouble[220:230])
print(quadDouble[220:230])
```

```python
inpX86, quadX86 = list(), list()
with open('iofiles/gmsk_float_x86_output.txt') as fX86:
    lines = fX86.readlines()
    for i in range(0, len(lines), 3): 
        line1 = lines[i].strip()
        line2 = lines[i + 1].strip()
        inpX86.append(float(line1))
        quadX86.append(float(line2))
print(inpX86[220:230])
print(quadX86[220:230])
```

```python
# float inPhase vs Quadrature graph
plt.plot(inpFloat, quadFloat)
plt.grid()
plt.show()
```

```python
# double inPhase vs Quadrature graph
plt.plot(inpDouble, quadDouble)
plt.grid()
plt.show()
```

```python
# x86 inPhase vs Quadrature graph
plt.plot(inpX86, quadX86)
plt.grid()
plt.show()
```

```python
# float inPhase 
plt.plot(inpFloat, 'r-')
plt.grid()
plt.show()
```

```python
# double inPhase
plt.plot(inpDouble, color=(1.0, 0.5, 0.0))
plt.grid()
plt.show()
```

```python
# x86 inPhase
plt.plot(inpX86[:500], color=(0, 0, 0))
plt.grid()
plt.show()
```

```python
# float Quadrature
plt.plot(quadFloat, 'b-')
plt.grid()
plt.show()
```

```python
# double quadrature
plt.plot(quadDouble, color=(1.0, 0.0, 1.0))
plt.grid()
plt.show()
```

```python
# x86 quadrature
plt.plot(quadX86[:500], color=(0, 0, 0))
plt.grid()
plt.show()
```
